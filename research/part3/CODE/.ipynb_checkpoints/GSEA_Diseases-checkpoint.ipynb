# Loading Files

```python
my_path = '/Users/deagogishvili/Documents/GitHub/hydrophobic_patches/part3/GSEA_RESULTS/Diseases'
file_path = '/Users/deagogishvili/Documents/GitHub/hydrophobic_patches/part3/FIGURES'
```

```python
THSA_neg = my_path + '/THSA_neg.tsv'
THSA_neg = pd.read_csv(THSA_neg, sep='\t', engine='python')
THSA_neg = THSA_neg[['NAME', 'ES', 'NOM p-val']]
THSA_neg = THSA_neg[THSA_neg['NOM p-val'] < 0.05]
THSA_neg = THSA_neg[THSA_neg['ES'] < -0.2]

THSA_pos = my_path + '/THSA_pos.tsv'
THSA_pos = pd.read_csv(THSA_pos, sep='\t', engine='python')
THSA_pos = THSA_pos[['NAME', 'ES', 'NOM p-val']]

THSA_pos = THSA_pos[THSA_pos['ES'] > 0.2]
THSA_pos['NOM p-val'] = THSA_pos['NOM p-val'].astype(float)
THSA_pos = THSA_pos[THSA_pos['NOM p-val'] < 0.05]

#Join positively and negatively enriched gene sets

frames = [THSA_pos, THSA_neg]
THSA = pd.concat(frames)
THSA = THSA.rename(columns = {'ES': 'ES_THSA'})
THSA
```

```python
RHSA_neg = my_path + '/RHSA_neg.tsv'
RHSA_neg = pd.read_csv(RHSA_neg, sep='\t', engine='python')
RHSA_neg = RHSA_neg[['NAME', 'ES', 'NOM p-val']]
RHSA_neg = RHSA_neg[RHSA_neg['NOM p-val'] < 0.05]
RHSA_neg = RHSA_neg[RHSA_neg['ES'] < -0.2]

RHSA_pos = my_path + '/RHSA_pos.tsv'
RHSA_pos = pd.read_csv(RHSA_pos, sep='\t', engine='python')
RHSA_pos = RHSA_pos[['NAME', 'ES', 'NOM p-val']]

RHSA_pos = RHSA_pos[RHSA_pos['ES'] > 0.2]
RHSA_pos['NOM p-val'] = RHSA_pos['NOM p-val'].astype(float)
RHSA_pos = RHSA_pos[RHSA_pos['NOM p-val'] < 0.05]

#Join positively and negatively enriched gene sets

frames = [RHSA_pos, RHSA_neg]
RHSA = pd.concat(frames)
RHSA = RHSA.rename(columns = {'ES': 'ES_RHSA'})
RHSA
```

```python
LHP_neg = my_path + '/LHP_neg.tsv'
LHP_neg = pd.read_csv(LHP_neg, sep='\t', engine='python')
LHP_neg = LHP_neg[['NAME', 'ES', 'NOM p-val']]
LHP_neg = LHP_neg[LHP_neg['NAME'] != 'ROSTY_CERVICAL_CANCER_PROLIFERATION_CLUSTER']
LHP_neg = LHP_neg[LHP_neg['ES'] < -0.2]
LHP_neg['NOM p-val'] = LHP_neg['NOM p-val'].astype(float)
LHP_neg = LHP_neg[LHP_neg['NOM p-val'] < 0.05]

LHP_pos = my_path + '/LHP_pos.tsv'
LHP_pos = pd.read_csv(LHP_pos, sep='\t', engine='python')
LHP_pos = LHP_pos[['NAME', 'ES', 'NOM p-val']]

LHP_pos = LHP_pos[LHP_pos['ES'] > 0.2]
LHP_pos['NOM p-val'] = LHP_pos['NOM p-val'].astype(float)
LHP_pos = LHP_pos[LHP_pos['NOM p-val'] < 0.05]

#Join positively and negatively enriched gene sets

frames = [LHP_pos, LHP_neg]
LHP = pd.concat(frames)
LHP = LHP.rename(columns = {'ES': 'ES_LHP'})
LHP
```

```python
GSEA = THSA.merge(RHSA, how='outer', on='NAME')
GSEA = GSEA.merge(LHP, how='outer', on='NAME')
```

```python
keep = GSEA.dropna(subset=['ES_RHSA', 'ES_LHP'], how='all')
keep = keep.dropna(subset=['ES_RHSA', 'ES_THSA'], how='all')
keep = keep.dropna(subset=['ES_LHP', 'ES_THSA'], how='all')

dfi.export(keep, file_path + '/GSEA_table.png')
keep
```

```python
keep = keep.sort_values('ES_RHSA', ascending=False)
s = keep.style\
    .background_gradient(cmap='coolwarm', subset=['ES_THSA'])\
    .background_gradient(cmap='coolwarm', subset=['ES_RHSA'])\
    .background_gradient(cmap='coolwarm', subset=['ES_LHP'])
dfi.export(s, file_path + '/GSEA_Diseases_Colored.png')
s
```
